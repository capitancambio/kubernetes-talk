```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
 
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D'],
'SQL': [38, 1.5, 30, 10],
'Machine Learning': [3, 20, 10, 40],
'Software Engineering': [8, 35, 2, 30],
'Python': [20, 35, 2, 30],
'Awesomeness': [30, 30, 30, 30]
})
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# We are going to plot the first line of the data frame.
# But we need to repeat the first value to close the circular graph:
def get_values(idx, values):
    values=df.loc[idx].drop('group').values.flatten().tolist()
    values += values[:1]
    return values
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories, color='grey', size=8)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)

colors = ['b','g', 'c', 'm']
for group in range(4):
    # Plot data
    ax.plot(angles, get_values(group, values), linewidth=1, linestyle='solid')
 
    # Fill area
    fill, = ax.fill(angles, get_values(group, values), colors[group], alpha=0.1)
    fill.set_label(f"Team member #{group}")
plt.legend(loc="lower right", bbox_to_anchor=(1.5, 0.8))
plt.show()
```

# Problem statement

_How to support growth (technically) with a heterogenous backlog, technologies, and team?_


* Give support to different type of processes: ETL, DL libraries, intensive CPU tasks...
* Different implementation approaches: python scripts, report generation, notebooks...
* Encouraging testability and repeatability.

# Airflow

# Airflow as central point of DS processes

_Airflow is a platform to programmatically author, schedule and monitor workflows._ 
[airflow.apache.org](http://airflow.apache.org)

* Glorified crontab.
* But has a (niceish) UI.
* Workflows are defined as DAGs (directed acyclic graphs of tasks) in pure python.

(Other alternatives lugi, argo, pure crontab, etc)

<center>
    <img src="img/dag.png">
</center>

# Airflow features

* Web UI
* Scheduler (with backfilling).
* XCOM (tasks can communicate amongst them).
* Already made hooks to get data from different sources.
* Tons of plug-ins.
* Allows to define your ETL logic in python along with the configuration of the DAG.

# Airflow features (that we actually use)

* Web UI
* Scheduler (with backfilling).
* ~~XCOM (tasks can communicate amongst them).~~ **Adds unnesary complexity to the system.**
* ~~Already made hooks to get data from different sources.~~ **Credentials are available to everone**
* ~~Tons of plug-ins.~~ **Everyone needs to be an Airflow expert**
* ~~Allows to define your ETL logic in python along with the configuration of the DAG.~~ **don't get me started**

I'll do it anyways:
* Testing incredibly difficult.
* Difficult maintainance: procedural definition tangled with complex logic.
* Dependecies are shared among all the dags. (2Gb of spacy models in your system to just multiply two matrices...)
* Using different hardware becomes painful (CUDA 👀) 

# Move everything out airflow 

* Pure python scripts that are enforced to follow a simple layered architecure.
  * Command line interface
  * Data loading layer
  * Transformation layer <- easy testing
  * Data storing layer
* Better testing and usable outside of airflow.
* Dockerise (containarise is a mouthful) the script.
   * Only the needed dependencies are installed.
   * Different versions can be used by different scripts.  
* Run it from airflow (_KubernetesPodOperator_)

# Kubernetes



Kubernetes is an open-source system for automating deployment, scaling, and management of containerized applications.
[kubernetes.io](https://kubernetes.io)

* Run our processes from the docker images (aka pods).
* Supports different types of hardware assigning the right pod to the right hardware (GPU, CPU, other resources, etc).
* Huge abstraction layer for cluster maintainace, secrets management, and resource management.
* Easy monitoring of tasks and memory usage.
* Active community.
* Supports your favourite cloud provider.

# Kubernetes (the downsides)

* Steep learning curve
* Thought for microservices rather than one-off tasks.
